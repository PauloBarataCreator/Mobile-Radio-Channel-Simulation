```python
from main import transmittion, mostrar_resultados
import matplotlib.pyplot as plt
```

## Setando Limiar

```python
LIMIAR = 200000
```

## Simulando Transmissão

```python
resultados = []
resultados = transmittion(LIMIAR, resultados)
```

## Resultados

### Resultado tabelado

```python
mostrar_resultados(resultados)
```

### Resultado Gráfico

```python
eb_no = [linha["Eb_No"] for linha in resultados]

plt.figure(figsize=(9, 6))
plt.semilogy(eb_no, [linha["BER_AWGN"] for linha in resultados], "o-", label="AWGN")
plt.semilogy(
    eb_no,
    [linha["BER_Rayleigh_AWGN"] for linha in resultados],
    "s-",
    label="Rayleigh plus AWGN",
)

plt.grid(True, which="both", linestyle=":")
plt.xlabel("Eb/No (dB)")
plt.ylabel("BER / Perro")
plt.title("Mobile Radio Channel Simulation")
plt.legend()
plt.tight_layout()

plt.show()
```
